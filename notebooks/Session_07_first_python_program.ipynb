## 💡 Program 1: Even or Odd Checker

We will ask the user for a number, check if it's even or odd, and show the result.

### 🔤 Pseudocode Recap

```

START
ASK user: "Enter a number"
STORE input in NUMBER
IF NUMBER modulo 2 equals 0 THEN
OUTPUT "Even number"
ELSE
OUTPUT "Odd number"
ENDIF
END

````

### 🐍 Python Code

```python
# Even or Odd Checker
number = int(input("Enter a number: "))

if number % 2 == 0:
    print("Even number")
else:
    print("Odd number")
````

✅ Run it, test it with different numbers, and observe the result.



## 🌡️ Program 2: Celsius to Fahrenheit Converter

We will convert temperature from Celsius to Fahrenheit.

### 🔤 Pseudocode Recap

```
START
ASK user: "Enter temperature in Celsius:"
STORE input in CELSIUS
CALCULATE FAHRENHEIT = (CELSIUS × 9/5) + 32
OUTPUT "Temperature in Fahrenheit is:", FAHRENHEIT
END
```

### 🐍 Python Code

```python
# Celsius to Fahrenheit Converter
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32

print("Temperature in Fahrenheit is:", fahrenheit)
```

🧪 Try it with values like `0`, `25`, or `100`.
