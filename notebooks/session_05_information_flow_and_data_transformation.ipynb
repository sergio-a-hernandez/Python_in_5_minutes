# Session 05 – Information Flow and Data Transformation



## 🔄 Understanding the Flow of Information

In every program, something comes **in**, something gets **done**, and something comes **out**.

This is called the **IPO Model**:

| Step      | Description                                 |
|-----------|---------------------------------------------|
| Input     | Data that comes from the user or another system |
| Process   | The logic or transformation applied to the data |
| Output    | The result shown to the user or stored somewhere |



## 🧠 Think Like a Programmer: Input → Process → Output

To write good code, you must think like a data processor.

- What data am I receiving?
- What must I do with it?
- What result should I give back?

Let’s take a real-world example.



## 🌡️ Example – Celsius to Fahrenheit Converter

We want to convert temperature from Celsius to Fahrenheit using the formula:

```

Fahrenheit = (Celsius × 9/5) + 32

```

Here’s how it looks with the IPO model:

**Input**
- Celsius temperature (from user)

**Process**
- Multiply Celsius by 9/5, then add 32

**Output**
- Fahrenheit temperature

---

## 🧾 Algorithm

1. Ask the user for temperature in Celsius
2. Apply the conversion formula
3. Show the result in Fahrenheit

---

## 🔤 Pseudocode

```

START
    ASK user: "Enter temperature in Celsius:"
    STORE input in CELSIUS
    CALCULATE FAHRENHEIT = (CELSIUS × 9/5) + 32
    OUTPUT "Temperature in Fahrenheit is:", FAHRENHEIT
END

```

## ✅ Summary

Every program is a transformation machine:
- **Input** is raw data
- **Process** is the logic or formula
- **Output** is the useful result

Understanding the IPO model helps you design better programs before writing any code.
